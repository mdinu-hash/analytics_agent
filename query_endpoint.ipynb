```python
from typing import Dict, Any
import os, requests
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()

def query_serving_endpoint(endpoint_name, user_message:str, thread_id: str = None) -> Dict[str,Any]:
    
   if thread_id:
        inputs = [
            {"role":"user","content":user_message},
            {"role":"system","content":f"thread_id:{thread_id}"}  # Fixed: removed space
        ]            
   else:
        inputs = [
            {"role":"user","content":user_message}
        ]

   try:
        # Use dataframe_records parameter instead of inputs
        response = w.serving_endpoints.query(
            name=endpoint_name,
            dataframe_records=[{"input": inputs}]  # Wrap inputs properly
        )
        return response
   except Exception as e:
        return {"error":str(e)}
```

```python
endpoint_name = 'analytics-agent-endpoint'  # Use your actual endpoint name
user_message = 'business model segments and their HNW number'

result = query_serving_endpoint(endpoint_name=endpoint_name, user_message=user_message)
print(result)
```
