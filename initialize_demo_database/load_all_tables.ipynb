```python
import os
import sqlite3
from dotenv import load_dotenv
from pathlib import Path
import sys

current_dir = Path.cwd()
root_path = current_dir.parent.parent.parent  # 3 levels up from init_demo_database_sqlite/
sys.path.insert(0, str(root_path))

env_path = root_path / '.env'
load_dotenv(env_path, override=True)

# Make db_path absolute so sqlite3 always resolves to project root
_raw = os.getenv('DB_PATH', './demo.db').lstrip('./')
db_path = str(root_path / _raw)
print(f'project root : {root_path}')
print(f'Using SQLite database: {db_path}')

# Run the schema creation script (creates all tables and indexes)
schema_script_path = root_path / 'src/init/init_demo_database_sqlite/a_create_tables.sql'
with open(schema_script_path, 'r', encoding='utf-8') as f:
    schema_sql = f.read()

conn = sqlite3.connect(db_path)
conn.executescript(schema_sql)
conn.close()
print('Schema initialized successfully (all tables created/verified)')
```

```python
# load date table

from src.init.init_demo_database_sqlite.b_create_date import DateDimensionGenerator

date_generator = DateDimensionGenerator(connection_string=db_path)

# Check if data exists and clear it
try:
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute('SELECT COUNT(*) FROM date')
    count = cursor.fetchone()[0]
    conn.close()
    if count > 0:
        print(f'Found {count} existing records in date table. Clearing...')
        conn = sqlite3.connect(db_path)
        conn.execute('DELETE FROM date')
        conn.commit()
        conn.close()
        print('Existing data cleared.')
except Exception as e:
    print(f'Note: Could not check/clear existing data: {e}')

# Generate the date dimension data
date_data = date_generator.generate_date_dimension_data()
print(f'Generated {len(date_data):,} date records')
print(f'Sample record: {date_data[0]}')

# Insert the data into the database
print('Inserting date dimension data into database...')
date_generator.insert_date_dimension_data(date_data)
print('Data insertion completed!')
```
