```python
# query a table
import sqlite3
import os
from pathlib import Path
current_dir = os.getcwd()
project_root = os.path.abspath(os.path.join(current_dir,".."))

conn = sqlite3.connect(str(Path(project_root) / 'demo.db'))
cursor = conn.cursor()
query = "SELECT distinct concat ('feedback dates between ',min(feedback_date),' and ', max(feedback_date) ) FROM fact_customer_feedback"
cursor.execute(query)
for row in cursor.fetchall():
    print(row)
conn.close()
```
